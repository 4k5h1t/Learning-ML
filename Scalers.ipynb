# 1. MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
X=np.array([[1,2,3,4,5,6,500],[-100,-50,0,50,100,2000,10000],[2,1,1,1,1,1,1]]) #3 Samples, 7 attributes or features
print(X)
print()
print(X.shape)
print()
scaler = MinMaxScaler()
scaler.fit(X)
xscale=scaler.transform(X)
print(xscale)
```

**Min Max Scaling**
$$xscale=\frac{x-x_{min}}{x_{max}-x_{min}}$$
**Example**

How 0.99019608 is obtained?

Values in first columns are $1,-100,2$. $x_{min}=-100$,$x_{max}=2$

Calculating xscale for first value, that is $x=1$
$$xscale=\frac{1-(-100)}{2-(-100)}$$
which means
$$xscale=\frac{101}{102}$$

```python
101/102
```

MinMaxScaler rescales the data set such that all feature values are in the range [0, 1]

A consequence of this is that all the feature values that are 'inliers' (ie. in the range of [0, 1]) will be within a very narrow space. 

# 2. RobustScaler

```python
from sklearn.preprocessing import RobustScaler
import numpy as np
X = np.array([[1,2,3,4,5,6,500],[-100,-50,0,50,100,2000,10000],[2,1,1,1,1,1,1]]) #3 Samples, 7 attributes or features
print("Original\n", X)
scaler = RobustScaler()
scaler.fit(X)
xscale = scaler.transform(X)
print("\n\nTransformed\n", xscale)
```


Robust Scaler algorithms are named so because, it scales features that are robust to outliers. The method it follows is almost similar to the MinMax Scaler but it uses the interquartile range (rather than the min-max used in MinMax Scaler). The median and scales of the data are removed by this scaling algorithm according to the quartile range.

To expand:

One approach to standardizing input variables in the presence of outliers is to ignore the outliers from the calculation of the mean and standard deviation, then use the calculated values to scale the variable.

This is called robust standardization or robust data scaling.

This can be achieved by calculating the median (50th percentile) and the 25th and 75th percentiles. The values of each variable then have their median subtracted and are divided by the interquartile range (IQR) which is the difference between the 75th and 25th percentiles.

It, thus, follows the following formula:

$$xscale=\frac{x-median(x)}{p75(x) - p25(x)}$$

  where p75 and p25 are the 75th and 25th percentile respectively


The resulting variable has a zero mean and median and a standard deviation of 1, although it is not skewed by outliers and the outliers are still present with the same relative relationships to other values.


Taking Feature 1, element 3:

$$xscale=\frac{2-1}{2 - (-100)}$$
$$xscale=\frac{1}{102}$$
$$xscale=0.0098$$


# 3. StandardScaler

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler().fit(X)
print("Original\n", X)
print("\n\nTransformed\n", ss.transform(X))
```

Standard Scaler follows the Standard Normal Distribution, i.e., it assumes a normal distribution for data within each feature. The scaling makes the distribution centered around 0, with a standard deviation of 1 and the mean removed.

For a data sample x, its score for Standard Scaler is calculated as:

$$xscale=\frac{x-mean(x)}{SD(x)}$$

where SD is Standard Deviation


Taking feature 1 : 

$$xscale=\frac{1-(-32.34)}{57.5}$$  


$$xscale = 0.62$$



# 4. Normalizer

```python
from sklearn.preprocessing import Normalizer

ss = Normalizer().fit(X)
print("Original\n", X)
print("\n\nTransformed\n", ss.transform(X))
```

The Normalizer rescales the vector for each sample to have unit norm, independently of the distribution of the samples. 

$$xscale=\frac{x}{norm(X)}$$  

$$xscale=\frac{1}{norm(1, 2, 3, 4, 5, 6, 500)}$$  

$$xscale=\frac{x}{500.09}$$

$$xscale=0.00199$$
